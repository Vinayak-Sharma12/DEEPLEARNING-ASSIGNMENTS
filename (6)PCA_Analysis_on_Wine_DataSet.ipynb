# **Principal Component Analysis(PCA)** -> It is a dimensionality reduction technique which keep important feature and it helps in improving the accuracy


Steps Required for PCA:
 1. Standardize the Data: Before Applying the PCA,ensure the data is scaled properly
 2. Apply PCA


# **IMPORT LIBRARIES**

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

# **LOAD DATASET**

```python
df=pd.read_csv('/content/winequality_red.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
#DROP THE TARGET COLOUMN
x=df.drop('quality',axis=1)
y=df['quality']
x.head()
```

```python
y.head()
```

# **APPLYING SCALING/STANDARDIZATION ON INPUT FEATURE**

Standardization(Scaling) refers to transforming the features so that they have a mean of 0 and a standard deviation of 1.

Each feature has mean of 0 and standard deviation of 1

z= σ/ x−μ
​


```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
#x_scaled will be matrix
```

```python
x_scaled
```

# **APPLY PCA**

```python
pca=PCA(n_components=2) #All feature will be reduced to 2 feature
X_pca=pca.fit_transform(x_scaled)
print(X_pca)
```

# **Visualize the Data**

```python
# Visualize the results
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA on Wine Dataset')
plt.colorbar(label='Class')
plt.show()
```
